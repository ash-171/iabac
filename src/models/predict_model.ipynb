### Import the saved model and check performance with sample data

```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('data\\external\\XGBClassifier_model.pkl', 'rb') as file:
    trained_model = pickle.load(file)
```

```python
dataset = pd.read_csv('data\\processed\\processed_data.csv')
```

```python
sample_size = 25 

random_sample = dataset.sample(n=sample_size, random_state=42)  

random_sample
```

```python
X = random_sample.drop(columns=['PerformanceRating'],axis=1)
```

```python
trained_model.predict(X)
```

```python
y = random_sample['PerformanceRating']
class_mapping = {label: idx for idx, label in enumerate(np.unique(y))}
y_mapped = np.vectorize(class_mapping.get)(y)
y_mapped
```

Saved saveral models.
* When `XGBClassifier` and `Random Forest` is used, it's predicting almost all of the values correctly.
